```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
from scipy import stats
```

# 데이터시각화 10/6일자 과제 201514142 김동준

```python
np.random.seed(201514142) 
x=np.random.chisquare(df=5, size=100)
```

```python
fig, ax =plt.subplots(1,3)
(ax1,ax2,ax3) =ax
```

```python
sns.boxplot(x,ax=ax1)  #ax1에 박스플랏을 넣겠다
sns.histplot(x,kde=True,ax=ax2) #ax2에 히스토그램
_ = stats.probplot(x,plot=ax3) #ax3에 qq플랏
```

```python
fig.set_figwidth(10)
fig.set_figheight(5)
fig.tight_layout()
fig
```
